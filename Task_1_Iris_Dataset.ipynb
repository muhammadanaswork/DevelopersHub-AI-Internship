# Task 1: Exploring and Visualizing a Simple Dataset
**Objective:** Load, inspect, and visualize the Iris dataset to understand data trends and distributions.
**Goal:** Practice basic data manipulation and exploratory data analysis (EDA) using Pandas, Matplotlib, and Seaborn.

```python
# Importing necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# We can load the Iris dataset directly from seaborn without downloading a CSV!
print("Loading dataset...")
df = sns.load_dataset('iris')
print("Dataset loaded successfully!")
```

### 1. Data Inspection and Summary Statistics
Here we will look at the shape of our data, the columns we have, and some basic statistical summaries.

```python
# Print the shape (number of rows and columns)
print("Dataset Shape:", df.shape)
print("\n")

# Print the column names
print("Column Names:", df.columns.tolist())
print("\n")

# Print the first few rows using .head()
print("First 5 rows of the dataset:")
display(df.head()) # 'display' makes it look nice in Jupyter/Colab
```

```python
# Use .info() to check data types and missing values
print("Dataset Info:")
df.info()
print("\n" + "="*50 + "\n")

# Use .describe() for summary statistics (mean, min, max, etc.)
print("Summary Statistics:")
display(df.describe())
```

### 2. Data Visualization
Visualizing the dataset to find relationships between features, distributions, and potential outliers.

```python
# Create a scatter plot to show relationships between features (Sepal Length vs Sepal Width)
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='sepal_length', y='sepal_width', hue='species', palette='viridis', s=100)
plt.title("Scatter Plot: Sepal Length vs Sepal Width")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Sepal Width (cm)")
plt.grid(True)
plt.show()
```

```python
# Use histograms to show value distributions for all numerical features
df.hist(figsize=(10, 8), bins=20, color='skyblue', edgecolor='black')
plt.suptitle("Histograms of Iris Features", fontsize=16)
plt.show()
```

```python
# Use box plots to identify outliers
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, orient="h", palette="Set2")
plt.title("Box Plots to Identify Outliers in Iris Features")
plt.xlabel("Centimeters (cm)")
plt.show()
```

### Final Insights and Explanation of Results
1. **Data Inspection:** The dataset contains 150 rows and 5 columns. There are no missing values, which makes preprocessing simple.
2. **Scatter Plot:** We can clearly see that the 'setosa' species is easily separable from the other two species based on sepal length and width.
3. **Histograms:** Petal length and petal width show a bimodal distribution (two peaks), which strongly hints at the presence of distinct species groups.
4. **Box Plots:** There are a few minor outliers in the `sepal_width` feature, but the data is largely clean and uniformly scaled.
